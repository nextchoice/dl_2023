```python
from keras.layers import Dense, Flatten
from keras.models import Sequential, Model
from keras.utils import to_categorical

from keras.datasets import mnist

import keras.models as models
import numpy as np

dir(models)
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train_trans = X_train.reshape(-1, X_train.shape[1]*X_train.shape[2])
y_train_trans = to_categorical(y_train)
model = Sequential()

model.add(Dense(256, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation="softmax"))

model.compile(optimizer='rmsprop', loss="categorical_crossentropy", metrics=['accuracy'])
model.fit(X_train_trans, y_train_trans, epochs=10, batch_size=100)
```

```python
from keras.losses import CategoricalCrossentropy
from keras.metrics import Accuracy, MSE
from keras.optimizers import Adam

model = Sequential(
    Dense(256, activation='relu',input_shape=(784,)),
    Dense(64, activation='sigmoid'),
    Dense(y_train_trans.shape[1], activation='softmax')
)

optimizer = Adam(learning_rate=1e-4)
metrics = ['accuracy']
loss = CategoricalCrossentropy

model.compile(optimizer=optimizer,
              loss=loss,
              metrics=metrics)
```
