```python
class Education():
    students = []
    admin_data = {'id':'admin', 'password':'1234'}
    user_data = {'id':''}
    def __init__(self, s):
        self.students = s
        #로그인 시작함.
        while True:
            self.login()
            if input('종료를 원하시면 "quit" 입력하세요.[계속 enter]') == 'quit':
                break
            
    def login(self):
        id = input("ID를 입력하세요.[학생은 이름을 입력하세요.]:::")
        pwd = input("비밀번호를 입력하세요.:::")
        self.user_data['id'] = id
        if id == self.admin_data['id'] and pwd == self.admin_data['password']:
            task_str = '1. 성적등록\n2. 성적삭제\n3.학생 성적 조회\n4.합계 및 평균 계산\n5.순위 조회\n6.전체 파일 저장'
            print(task_str)
            while True:
                self.display_admin_menu()
                if input('종료를 원하시면 "x" 입력하세요.[계속 enter]') == 'x':
                    break
            return True
        else:
            task_str = '1.학생 성적 조회\n2.순위 조회'
            print(task_str)
            while True:
                self.display_student_menu()
                if input('종료를 원하시면 "x" 입력하세요.[계속 enter]') == 'x':
                    break
            return True
        
    def display_student_menu(self):
        self.task = input('원하는 메뉴를 선택하세요::::')
        if self.task == '1':
            self.search_student(self.user_data['id'])
        elif self.task == '2':
            target = input("순위를 원하는 과목을 입력하세요.:::")
            self.rank_score(target, self.user_data['id'])
        else:
            print("선택한 메뉴가 정확하지 않습니다.[1~2 만 입력하세요.]")
        
    def  display_admin_menu(self):
        self.task = input('원하는 메뉴를 선택하세요::::')
        if self.task == '1':
            self.manage('new')
        elif self.task == '2':
            self.manage('del')
        elif self.task == '3':
            self.manage('name')
        elif self.task == '4':
            self.cal_score()
        elif self.task == '5':
            target = input("순위를 원하는 과목을 입력하세요.:::")
            self.rank_score(target)
        elif self.task == '6':
            self.manage('save')
        else:
            print("선택한 메뉴가 정확하지 않습니다.[1~6 만 입력하세요.]")
    
    def cal_score(self):
        for i, s in enumerate(self.students):
            total = 0
            cnt = 0
            for v in s.values():
                if isinstance(v, int):
                    total += v
                    cnt += 1

            if cnt > 0:
                s['total'] = float(total)
                s['avg'] = round(total / cnt, 2)
            else:
                s['total'] = total
                s['avg'] = total
                    
        k_total = [s['korean'] for s in self.students]
        m_total = [s['math'] for s in self.students]
        e_total = [s['english'] for s in self.students]
        t_total = [s['total'] for s in self.students]
        a_total = [s['avg'] for s in self.students]
        self.students.append({'name':'합계or평균', 'korean':round(sum(k_total)/len(k_total), 2),'math':round(sum(m_total)/len(m_total),2),
                              'english':round(sum(e_total)/len(e_total),2),'total':round(sum(t_total),2),
                              'avg':round(sum(a_total)/len(a_total),2)})
        print("계산이 완료 되었습니다.")
                
    def print_data(self):
        for s in self.students:
            print(f'이름 : {s["name"]} :::: 국어 : {s["korean"]} :::: 수학 : {s["math"]} :::: 영어 : {s["english"]}')
            
    def search_student(self, name):
        s_list = [s for s in self.students if s['name'] == name]
        for v in s_list:
            print(v)
        return s_list
    
    def search_score(self, subject, score):
        s_list = [s for s in self.students if s[subject] >= score]
        for v in s_list:
            print(v)
        print(f"찾은 내역 :: {s_list}")
        return s_list
    
    def rank_score(self, target, name=None):
        rank_list = sorted(self.students, key=lambda d: d[target], reverse=True)
        
        if name != None:
            for i, v in enumerate(rank_list):
                if v['name'] == name:
                    print(f"{v['name']} 님의 {target} 과목의 순위는 {i+1}등 입니다.")
        else:
            print(f"전체 순위 Data :: {rank_list}")
            
        return rank_list
    
    def manage(self,method):
        if method == 'del':
            name = input('삭제할 학생의 이름을 입력하세요.:::')
            self.delete(name)
            return True
        
        if method == 'new':
            self.create()            
            return True
        
        if method == 'name':
            name = input('조회를 원하는 학생명을 입력하세요.:::')
            self.search_student(name)
            return True
            
        if method == 'save':
            self.save()
            return True
          
        print('지원하는 method가 없습니다.')        
        return False

    def save(self):
        f_df = pd.DataFrame.from_dict(self.students)
        f_df.to_csv(r'education_st_list.csv', index=False, header=True)
        print("저장처리가 되었습니다.")
        
    def delete(self, name):
        self.students = list(filter(lambda d: d['name'] != name, self.students))
        print("List after deleting of dictionary :::" + str(self.students))
        
    def create(self):
        while(True):
            name = input('입력할 학생의 이름을 입력하세요:::')
            k = input('국어점수를 입력하세요:::')
            m = input('수학점수를 입력하세요:::')
            e = input('영어점수를 입력하세요:::')
            d = {'name':name,'korean':int(k),'math':int(m),'english':int(e)}
            self.students.append(d)
            print(f"{name} 입력되었습니다.")
            if input('종료를 원하시면 "q" 를 입력하세요:::[계속:enter]') == 'q':
                break
```
