```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
#한글 깨짐 방지
plt.rcParams['font.family'] = 'NanumGothic'
#마이너스 부호 깨짐 방지
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = np.arange(1, 10, 0.5)
y = x * 10
plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('제목입니다.')
plt.show()
```
